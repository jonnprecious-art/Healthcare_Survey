```python
import pandas as pd

df = pd.read_csv('users.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
import matplotlib.pyplot as plt

# Sum income by age
age_income = df.groupby('Age')['Income'].sum().sort_values(ascending=False)

# Plot
age_income.plot(kind='bar', figsize=(10,6))
plt.title("Ages with Highest Income")
plt.ylabel("Total Income")
plt.xlabel("Age")
plt.show()
```

```python
df.groupby('Gender')['Shopping'].sum().plot(kind='pie', autopct='%1.1f%%', figsize=(6,6))
plt.title("Shopping by Gender")
plt.show()
```
